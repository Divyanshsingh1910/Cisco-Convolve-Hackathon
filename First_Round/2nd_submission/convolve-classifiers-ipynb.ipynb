```python
import numpy as np
import pandas as pd
```

```python
import json
with open('/kaggle/input/convolve-epoch1/train.json','r') as f:
    log=json.load(f)
    
data = pd.DataFrame({'Label':log})
```

```python
data
```

```python
arr1=np.array(data.index)
arr2=np.array(data['Label'])
train=pd.DataFrame()
train['Log']=arr1
train['Label']=arr2
train
```

```python
train_abnormal = train[train['Label']=='abnormal']
train_abnormal
```

```python
train_normal = train[train['Label']=='normal']
train_normal
```

```python
train_normal_downsampled = train_normal.sample(2*train_abnormal.shape[0])
train_normal_downsampled.shape
```

```python
train = pd.concat([train_normal_downsampled,train_abnormal])
train
```

```python
train.reset_index(inplace=True)
train
```

```python
train.drop('index',axis=1, inplace=True)
train
```

```python
train['Label'].value_counts()
```

```python
import random
for i in range(5):
    index = random.randint(0,len(train['Log']))
    print(str(index)+'th Log is: '+train['Log'][index])
```

```python
#now we will clean n data a bit
import nltk
import re
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
wordnet=WordNetLemmatizer()
```

```python
for i,sentence in enumerate(train['Log']):
    new_sentence = re.sub('[^a-zA-Z]', ' ', sentence)
    train['Log'][i]=new_sentence
```

```python
for i in range(5):
    index = random.randint(0,len(train['Log']))
    print(str(index)+'th Log is: '+train['Log'][index])
```

```python
import nltk
nltk.download('omw-1.4')
```

```python
corpus = []
for i in range(len(train['Log'])):
    review = train['Log'][i].lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len(corpus)==len(train['Log'])
```

```python
#Let's checkout some lemmatized texts
for i in range(5):
    index = random.randint(0,len(train['Log']))
    print(str(index)+'th Log is: '+corpus[index])
```

```python
#let's vectorize them
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
train_Lemmatized=pd.DataFrame(corpus,columns={'Log'})
train_Lemmatized
```

```python
train_Lemmatized.to_csv('/kaggle/working/train_Lemmatized.csv')
```

```python
transformer =TfidfVectorizer(max_features=500)
X = transformer.fit_transform(corpus).toarray()
```

```python
X
```

```python
X.shape
```

```python
np.unique(X[0])
```

```python
Y = train['Label'].apply(lambda x: 1 if x=='normal' else 0)
Y
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X,Y)
model.score(X,Y)
```

```python
y_predicted = model.predict(X)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(Y, y_predicted)
cm 
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier()
model_rf.fit(X,Y)
model_rf.score(X,Y)
```

```python
y_predicted = model_rf.predict(X)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(Y, y_predicted)
cm 
```
