```python
import json
with open('/content/drive/MyDrive/Convolve IITG/train.json','r') as f:
    data=json.load(f)
    
train = pd.DataFrame({'Label':data})
```

```python
more_ab_data=pd.read_csv("/content/drive/MyDrive/Convolve IITG/more_abnomal.csv")
more_ab_data.drop('ID',axis=1,inplace=True)
more_ab_data
more_ab_data.rename(columns = {' Log':'Log'}, inplace = True)
```

```python
arr1=np.array(train.index)
arr2=np.array(train['Label'])
train=pd.DataFrame()
train['Log']=arr1
train['Label']=arr2
train
```

```python
train_abnormal = train[train['Label']=='abnormal']
train_abnormal
```

```python
train_abnormal_downsampled = train_abnormal.sample(69692)
train_abnormal_downsampled.columns
```

```python
train_normal = train[train['Label']=='normal']
train_normal.shape
```

```python
#train_normal_downsampled = train_normal.sample(train_abnormal.shape[0])
train_normal_downsampled = train_normal.sample(500000)
train_normal_downsampled.columns
```

```python
def func(x):
    count_space=0
    index=0
    for i in range(len(x)):
        if x[i]==' ':
            count_space = count_space+1
        if count_space==6:
            index=i;
            break
    return x[index+1:]
```

```python
train_balance= pd.concat([train_normal_downsampled, train_abnormal_downsampled])
train_balance
```

```python
train_balanced = pd.concat([train_balance, more_ab_data])
train_balanced
```

```python
arr1=np.array(train_balanced['Log'])
arr1
```

```python
for i in range(train_balanced.shape[0]):
    arr1[i]=func(arr1[i])
arr1
```

```python
train_balanced['Log']=arr1
train_balanced
```

```python
from sklearn.utils import shuffle
shuffle(train_balanced, random_state=0)
```

```python
#train_balanced.reset_index(inplace=True)
#train_balanced
```

```python
train_balanced['normal']=train_balanced['Label'].apply(lambda x:1 if x=='normal' else 0)
train_balanced.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_balanced['Log'],train_balanced['normal'],stratify=train_balanced['normal'])
```

```python
bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")
```

```python
#Layers
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)
outputs = bert_encoder(preprocessed_text)

normalizer = tf.keras.layers.Normalization(axis=-1)

# Neural network layers
l = tf.keras.layers.Dropout(0.1, name="dropout")(outputs['pooled_output'])
#l = tf.keras.layers.Dense(200, activation = 'relu')(l)
#l = tf.keras.layers.Dense(50, activation = 'relu')(l)
l = tf.keras.layers.Dense(15, activation = 'tanh')(l)
l = tf.keras.layers.Dense(5, activation = 'tanh')(l)
l = tf.keras.layers.Dense(1, activation='sigmoid', name="output")(l)

# Use inputs and outputs to construct a final model
model = tf.keras.Model(inputs=[text_input], outputs = [l])
```

```python
model.summary()
```

```python
METRICS = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall')
]

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=METRICS)
```

```python
from tensorflow import keras
```

```python
model.fit(X_train, y_train, epochs=1)
```
